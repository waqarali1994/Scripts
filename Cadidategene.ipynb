```python
def ld_parse(afile):
    fh = open(afile)
    fh.readline()
    ld_dict = {}
    for x in fh:
        y = x.split()
        # print(y)
        ld_dict[y[-2]] = [float(y[-1]), y[-3]]
    return ld_dict
```
